```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(1)
```

```python
#1000 random integer between 0 and 50 
x=np.random.randint(0,50,1000)
```

```python
#positive corrrelation with some noise
y=x-np.random.normal(0,10,1000)
#positive correlation with some noise
y=x-np.random.normal(0,10,1000)
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
#negative correlation
x=np.random.randint(0,50,1000)
y=100-x-np.random.normal(0,5,1000)

```

```python
np.corrcoef(x,y)
```

```python

plt.scatter(x,y)
```

```python
x=np.random.randint(0,50,1000)
y=np.random.randint(0,50,1000)
plt.scatter(x,y)
```

```python
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a'] + np.random.normal(0,10,1000)#positive corelation
df['c']=100-df['a']+np.random.normal(0,5,1000)#negative corelation
df['d']=np.random.randint(0,50,1000)#-ve corelation with a
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
